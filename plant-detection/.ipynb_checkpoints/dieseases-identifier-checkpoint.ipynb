```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = 'DATASET/train'
data_test_path = 'DATASET/test'
data_val_path = 'DATASET/valid'
```

```python
img_width = 180
img_height = 180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False
)
```

```python
data_category = data_train.class_names
```

```python
data_category
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    image_size=(img_height,img_width),
    batch_size=32,
    shuffle=False,
    validation_split=False
)
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    image_size=(img_height,img_width),
    shuffle=False,
    batch_size=32,
    validation_split=False
)
```

```python
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9): # category 
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_category[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_category))
                  
])
```

```python
model.compile(
    optimizer='adam', 
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
    metrics=['accuracy']
)
```

```python
epochs_size = 25 #25
history = model.fit(data_train, validation_data=data_val, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```

```python
image = 'Healthy-tomato.jpg'
image = tf.keras.utils.load_img(image, target_size=(img_height,img_width))
img_arr = tf.keras.utils.array_to_img(image)
img_bat=tf.expand_dims(img_arr,0)
```

```python
predict = model.predict(img_bat)
```

```python
score = tf.nn.softmax(predict)
```

```python
category_class = data_category[np.argmax(score)]

if category_class == 'Healthy_tomato':
    print('This plant is Healthy with accuracy of {:0.2f}'.format(np.max(score)*100))
if category_class == 'Early_blight_tomato':
    print('This plant has this Early Blight disease with accuracy of {:0.2f}'.format(np.max(score)*100))
if category_class == 'Powdery_mildew_tomato':
    print('This plant has this Powdery Mildew disease with accuracy of {:0.2f}'.format(np.max(score)*100))
if category_class == 'Yellow_leaf_curl_virus_tomato':
    print('This plant has this Yellow leaf curl virus disease with accuracy of {:0.2f}'.format(np.max(score)*100))
if category_class != 'Yellow_leaf_curl_virus_tomato' or 'Powdery_mildew_tomato' or 'Early_blight_tomato' or 'Healthy_tomato':
    print('This is not Tomato plant')

# print(category_class)
```
